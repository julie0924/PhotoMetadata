```python
img = df['image'].to_dict()

names = []
for i in range(0, len(img)):
    names.append(img[i]['fileName'])
    
orientations =[]
for i in range(0, len(img)):
    if img[i]['details']['width'] < img[i]['details']['height']:
        orientation = 'portait'
        orientations.append(orientation)
    else: 
        orientation = 'landscape'
        orientations.append(orientation)

data = pd.DataFrame({'File_Name': names, 'Orientation': orientations})
data
```

```python
obj = df['objects'].to_dict()
obj[0][0]['bndbox']
obj[0][0]['category']['name']
```

```python
ls_name = []
ls_coordinate = []
for i in range(len(obj)):
    category_names = []
    bndbox_tuples = []

    for j in range(len(obj[i])):
        category_names.append(obj[i][j]['category']['name'])
        bndbox_dict = obj[i][j]['bndbox']
        bndbox_tuple = (bndbox_dict['minX'], bndbox_dict['minY'], bndbox_dict['maxX'], bndbox_dict['maxY'])
        bndbox_tuples.append(bndbox_tuple)

    ls_name.append(category_names)
    ls_coordinate.append(bndbox_tuples)
    
ls_name
ls_coordinate

ls_df = pd.DataFrame({'File_Name': names, 'bndbox_name': ls_name, 'coordinates': ls_coordinate})
ls_df
# list_df = pd.DataFrame({'bndbox_name': category_name, 'coordinates': bndbox_tuple})
# list_df

ls_df = ls_df.explode(['bndbox_name','coordinates'])
ls_df.head()

#file = file.explode('MKIDS', ignore_index = True)

#data = pd.DataFrame({'File_Name': names, 'Orientation': orientations})
```

```python
# importing json files
def get_json(file):
    file = './JSONforZB/'+ file
    f = open(file)
    data = json.load(f)
    return data

# List of JSON files
jsv1 = os.listdir('JSONforZB')
# List comprehension with the function inside, stored inside a variable
jsv2 = [get_json(i) for i in jsv1]

len(jsv2)
# df_ZB = pd.DataFrame(jsv2)
# # # image = jsv2['image'].to_dict()
# # # image
# df_ZB
# #df_ZB = pd.read_json(jsv2[0])
# # c
# # df_ZB
# jsv2

# new = jsv2[0]
# new
# # [[new_list]] = [[[jsv2]]]
# # print(new_list)
# jsv2


# list1 = [10, 11, 12, 13, 14] 
# list2 = [20, 30, 42] 


# res = list1 + list2 


# print ("Concatenated list:\n" + str(res)) 

conc_list = []

for i in range(0,len(jsv2)):
    conc_list = conc_list + jsv2[i]
    
conc_list

df_ZB = pd.DataFrame(conc_list)
df_ZB
```
